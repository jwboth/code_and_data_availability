## Results of the CompGeo analysis (imaging vs simulation)

```python
input_stem = "compgeo_img_vs_sim"
categories = ["imaging", "simulation"]
```

```python
!python ..\scripts\plotting.py -i ..\results\{input_stem}.csv -c {" ".join(categories)}
```

```python
from IPython.display import Image, display
import os

# Define the input file stem and output folder
output_folder = "results"

# List of expected figure suffixes
figure_suffixes = [
    "category_distribution.png",
]

# Add per-category pie chart suffixes
for cat in categories:
    figure_suffixes.append(f"paper_availability_pies_{cat}.png")
    figure_suffixes.append(f"data_availability_pies_{cat}.png")

# Display each figure inline
for suffix in figure_suffixes:
    fig_path = os.path.join(output_folder, f"{input_stem}_{suffix}")
    if os.path.exists(fig_path):
        display(Image(filename=fig_path))
    else:
        print(f"Figure not found: {fig_path}")
```
